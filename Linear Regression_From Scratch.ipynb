### Import Necessary Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Load the Data

```python
df = pd.read_csv('Depression and  Happiness Factor Analysis_1604050.csv')
df.head()
```

### Renaming the Columns

```python
df.columns = ['Education', 'Emotion_category','Scale','Gender','Age','Location','Relationship_status','Financial_state',' cope_with_environment','Understanding_with_family',
              'Feeling_in_study_work','Academic_result','Living_place','Who_supports?','social_media_activity','Inferiority_complex?','Todays_meal','Health_issues',
            'Recreational_activity','Sleeping_last_night']

df.head()
```

### Label Encoding the Categorical Features

```python
from sklearn.preprocessing import LabelEncoder
categorical_cols =['Education', 'Emotion_category','Gender', 'Location', 'Relationship_status','Financial_state','Understanding_with_family',
                     'Feeling_in_study_work','Academic_result','Living_place','Who_supports?','social_media_activity','Inferiority_complex?',
                     'Todays_meal','Health_issues','Recreational_activity'] 

le = LabelEncoder()
df[categorical_cols] = df[categorical_cols].apply(le.fit_transform)
df.head()
```

```python
# Rearrange
take_column = df.pop('Scale')
df.insert(0, 'Scale', take_column)
df.head()
```

### Separating Independent and Dependent Variables

```python
X=df.iloc[:,1:].values
Y= df.iloc[:,0].values
```

```python
print(X.shape)
print(Y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.2, random_state=0)
```

```python
print(X_train.shape)
print(Y_train.shape)
```

### Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler
scale_X=StandardScaler()   
X_train=scale_X.fit_transform(X_train)
X_test=scale_X.transform(X_test)
```

```python
X_train
```

### Linear Regression Class

```python
class Multiple_Linear_Regression:

    def __init__(self, learning_rate=0.001, n_iterations=1000):
        self.lr = learning_rate
        self.n_iter = n_iterations
        self.weights=None
        self.bias = None
        self.loss = []
        
    
    def fit(self, X, y):
        # Initialize weights and bias to zeros
        self.weights = np.zeros(X.shape[1])
        self.bias = 0
        
        # Perform Gradient Descent
        for i in range(self.n_iter):
            # Line equation
            y_hat = np.dot(X, self.weights) + self.bias
            loss = self.MSE(y, y_hat)
            self.loss.append(loss)
            
            # Calculate derivatives during back propagation
            dw = (1 / X.shape[0]) * (2 * np.dot(X.T, (y_hat - y)))
            db = (1 / X.shape[0]) * (2 * np.sum(y_hat - y))
            
            # Update the coefficients
            self.weights -= self.lr *dw 
            self.bias -= self.lr * db
            
    @staticmethod
    def MSE(y, y_hat):
        error = 0
        for i in range(len(y)):
            error += (y[i] - y_hat[i]) ** 2
        return error / len(y)
        
        
    def predict(self, X):
        return np.dot(X, self.weights) + self.bias
```

### Fit the Model on Training Data

```python
model= Multiple_Linear_Regression(learning_rate=0.01, n_iterations=20000)
model.fit(X_train, Y_train)
```

### Predict on Test Data


```python
y_preds = model.predict(X_test)
y_preds
```

### Check the R Square (R2) Score

```python
def R2 (y_preds, y):
    ssr = sum((y - y_preds) ** 2) 
    sst = sum((y - y.mean()) ** 2) 
    return 1 - ssr/sst
```

```python
  print('R Square Score: %.5f'% R2(y_preds ,Y_test))
```

### Visualizing the True and Predicted values 

```python
fig, ax = plt.subplots(figsize=(10,8))
ax.scatter(Y_test, y_preds)
ax.set_xlabel('True Values')
ax.set_ylabel('Fitted Values')
ax.set_title('True Value vs. Fitted Values')
```

### Check the performance against the built-in library provided by Scikit-learn 

```python
def Linear_Regression_scikitlearn(x_train, y_train, x_test, y_test):
    
    from sklearn.linear_model import LinearRegression
    from sklearn.metrics import r2_score
    
    model=LinearRegression()
    model.fit(x_train, y_train)
    
    y_pred=model.predict(x_test)

    print('R Square Score: %.5f'% r2_score(y_test, y_pred))
```
