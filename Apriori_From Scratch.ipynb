### Import the required libraries

### Load the Data

```python
df = pd.read_csv('Depression and  Happiness Factor Analysis_1604050.csv')
df.head()
```

### Renaming the Columns

```python
df.columns = ['Education', 'Emotion_category','Scale','Gender','Age','Location','Relationship_status','Financial_state',' cope_with_environment','Understanding_with_family',
              'Feeling_in_study_work','Academic_result','Living_place','Who_supports?','social_media_activity','Inferiority_complex?','Todays_meal','Health_issues',
            'Recreational_activity','Sleeping_last_night']

df.head()
```

### Splitting the Data According to Education Year



```python
basket_1st_year = (df[df['Education'] =="1st year"]
        .groupby(['Emotion_category', 'Location'])['Scale']
        .sum().unstack().reset_index().fillna(0)
        .set_index('Emotion_category'))
basket_1st_year 
```

```python
basket_2nd_year = (df[df['Education'] =="2nd year"]
        .groupby(['Emotion_category', 'Location'])['Scale']
        .sum().unstack().reset_index().fillna(0)
        .set_index('Emotion_category'))
basket_2nd_year  
```

```python
basket_3rd_year = (df[df['Education'] =="3rd year"]
        .groupby(['Emotion_category', 'Location'])['Scale']
        .sum().unstack().reset_index().fillna(0)
        .set_index('Emotion_category'))
basket_3rd_year
```

```python
basket_4th_year = (df[df['Education'] =="4th year"]
        .groupby(['Emotion_category', 'Location'])['Scale']
        .sum().unstack().reset_index().fillna(0)
        .set_index('Emotion_category'))
basket_4th_year 
```

```python
basket_Not_applicable = (df[df['Education'] =="Not applicable"]
        .groupby(['Emotion_category', 'Location'])['Scale']
        .sum().unstack().reset_index().fillna(0)
        .set_index('Emotion_category'))
basket_Not_applicable
```

### Encoding

```python
def encode_values(x):
    if x <= 0:
        return 0
    if x >= 1:
        return 1
basket_sets_1st = basket_1st_year.applymap(encode_values)
basket_sets_2nd = basket_2nd_year.applymap(encode_values)
basket_sets_3rd = basket_3rd_year.applymap(encode_values)
basket_sets_4th = basket_4th_year.applymap(encode_values)
basket_sets_not = basket_Not_applicable.applymap(encode_values)
```

### Apriori Algorithm to Generate Frequent Item Sets

```python
from itertools import combinations

def Apriori_Algorithm (data, min_support=0.02, max_length = 2):
    support = {} 
    L= list(data.columns)
    
    for i in range(1, max_length+1):
        comb = set(combinations(L,i))
        
        L =set()     
        for j in list(comb):
            sup = data.loc[:,j].product(axis=1).sum()/len(data.index)
            
            if sup > min_support:
                support[j] = sup
                L = list(set(L) | set(j))
        
    itemset = pd.DataFrame(list(support.items()), columns = ["Items", "Support"])
    return(itemset)
```

```python
frequent_item_sets_1st = Apriori_Algorithm (basket_sets_1st, 0.04, 3)
frequent_item_sets_1st.sort_values(by = 'Support', ascending =False)
```

```python
frequent_item_sets_2nd = Apriori_Algorithm (basket_sets_2nd, 0.04, 3)
frequent_item_sets_2nd.sort_values(by = 'Support', ascending =False)
```

```python
frequent_item_sets_3rd = Apriori_Algorithm (basket_sets_3rd, 0.04, 3)
frequent_item_sets_3rd.sort_values(by = 'Support', ascending =False)
```

```python
frequent_item_sets_4th = Apriori_Algorithm (basket_sets_4th, 0.04, 3)
frequent_item_sets_4th.sort_values(by = 'Support', ascending =False)
```

```python
frequent_item_sets_not = Apriori_Algorithm (basket_sets_not, 0.04, 3)
frequent_item_sets_not.sort_values(by = 'Support', ascending =False)
```

### Generate Association Rules Using Frequent Item Sets

```python
from itertools import permutations
def Generate_Association_Rules (df, min_threshold=0.5):
    support = pd.Series(df.Support.values, index=df.Items).to_dict()
    data = []
    L= df.Items.values
    perm = list(permutations(L, 2))
    
    # Iterating through each rule
    for i in perm:
        if set(i[0]).issubset(i[1]):
            conf = support[i[1]]/support[i[0]]
            
            if conf > min_threshold:
                j = i[1][not i[1].index(i[0][0])]
                lift = support[i[1]]/(support[i[0]]* support[(j,)])
                leverage = support[i[1]] - (support[i[0]]* support[(j,)])
                data.append([i[0], (j,), support[i[0]], support[(j,)], support[i[1]], conf, lift, leverage])

         

    out = pd.DataFrame(data, columns = ["antecedents", "consequents", "antecedent support", "consequent support",
                                        "support", "confidence", "Lift", "Leverage"])
    return(out)
```

```python
rules_not = Generate_Association_Rules (frequent_item_sets_not, 0.2)
rules_not
```
